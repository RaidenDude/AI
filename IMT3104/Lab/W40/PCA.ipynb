```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load iris dataset
iris = datasets.load_iris()

# Pandas to plot the data in a table
features = [ "Petal length","Petal Width","Sepal Length","Sepal Width" ]
data=pd.DataFrame(iris['data'],columns=features)
data['Species']=iris['target']
data['Species']=data['Species'].apply(lambda x: iris['target_names'][x])
data.head(10)
```

```python
X = data.loc[:, features].values
y = data.loc[:, ['Species']].values

# Normalize the data
scaler = StandardScaler()
X = scaler.fit_transform(X)
pdX = pd.DataFrame(X)
pdX.head(10)
```

```python
# PCA for dim.reduction 4->2
pca = PCA(n_components=2)
pca.fit(X)
X = pca.transform(X)

data=pd.DataFrame(X)
data.head(10)
```

```python
# Plot principal component 1/2 as a 2D graph
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
# Try to repeat PCA for 2->1 components, and plot the data
pca = PCA(n_components=1)
pca.fit(X)
X = pca.transform(X)

data=pd.DataFrame(X)
data.head(10)

y = np.zeros(len(X))

plt.scatter(X, y)
plt.show()
```
